```
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import concatenate,Input
from tensorflow.keras.utils import plot_model
import keras
from tensorflow.keras.layers import Dense,Activation,MaxPool2D,Flatten,Conv2D
```

```
# input layer
input_data=Input(shape=(28,28,1))
```

```
# first feature extractors
conv1=Conv2D(32,kernel_size=(5,5),activation='relu')(input_data)
pool1=MaxPool2D(pool_size=(2,2))(conv1)
flat1=Flatten()(pool1)
```

```
# second featuer extractors
conv2=Conv2D(32,kernel_size=(5,5),activation='relu')(input_data)
pool2=MaxPool2D(pool_size=(2,2))(conv2)
flat2=Flatten()(pool2)

```

```
# merge feature extractors
merge=concatenate([flat1,flat2])
```

```
#interpretation layer
hidden1=Dense(10,activation='relu')(merge)
```

```
#output layer
output=Dense(1,activation='sigmoid')(hidden1)
```

```
#model generation
model=Model(inputs=input_data,outputs=output)
```

```
#plot graph
plot_model(model,show_shapes=True,show_layer_names=True)
```
