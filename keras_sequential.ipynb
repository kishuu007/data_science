```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
from tensorflow.keras.utils import plot_model
import keras
```

```
model=Sequential()
model.add(Dense(8,activation='relu',input_shape=(2,)))
model.add(Dense(2,activation='softmax'))
```

```
model.summary()
```

```
plot_model(model,show_shapes=True,show_layer_names=True)
```

```
from tensorflow.keras.layers import Dense,Activation,MaxPool2D,Flatten,Conv2D
```

```
model=Sequential()
model.add(Conv2D(filters=32,kernel_size=(5,5),activation='relu', input_shape=(28,28,1)))
#output = 28-5+1=24 and 32 is no of nueral nodes
model.add(Conv2D(filters=32,kernel_size=(5,5),activation='relu'))
#output = 24-5+1=20
model.add(MaxPool2D(pool_size=(2,2)))
#  maxpool output shape will be  half of the input
model.add(Flatten())
# flatten output is input shape is flattened to single dimension i.e multi dim to single dim
model.add(Dense(2,activation='softmax'))
```

```
model.summary()
```

```
plot_model(model,show_shapes=True,show_layer_names=True)
```
